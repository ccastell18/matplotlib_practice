## Matplotlib example workflow

```python
#imoprt matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup
fig, ax = plt.subplots(figsize=(10, 10))

#3. Plot data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot",
       xlabel='x-axis',
       ylabel="y-axis")

#Save and show (you save the whole figure)
```

## Making figures with NumPy arrays

* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
#Plot data for line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
#same data for Scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
#plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter":12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax= plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
       xlabel="Price ($)");
```

```python
#make some data
x = np.random.randn(1000)
ax = plt.subplot()
ax.hist(x);
```

# Two options for subplots

```python
#Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

#Plot to each data
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

#plot each different
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].barh(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(x);
```

# Plotting from Pandas Dataframes

```python
car_sales = pd.read_csv('car-sales.csv')
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2024', periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].astype(str).str.replace('[\$\,\.]','',regex=True).astype(int)
```

```python
car_sales
```

```python
#Remove 00
car_sales["Price"] = pd.to_numeric(car_sales["Price"])
car_sales
```

```python
car_sales['Price'].dtype
```

```python
car_sales["Price"] = car_sales["Price"]/ 100
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2024', periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
#Let's plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales");
```
